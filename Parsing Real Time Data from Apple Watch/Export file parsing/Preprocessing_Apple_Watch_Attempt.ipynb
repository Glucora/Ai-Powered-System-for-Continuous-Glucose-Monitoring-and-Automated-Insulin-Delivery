```python
import pandas as pd

df = pd.read_csv('apple_health_merged.csv')
print("Data loaded successfully. Displaying the first 5 rows:")
df.head()
```

```python
df = df.dropna()
print("DataFrame shape after removing missing values:", df.shape)
```

```python
missing_values_count = df.isnull().sum()
missing_values_percentage = 100 * df.isnull().sum() / len(df)

missing_values_df = pd.DataFrame({
    'Missing Count': missing_values_count,
    'Missing Percentage': missing_values_percentage
})

missing_values_df = missing_values_df.sort_values(by='Missing Percentage', ascending=False)

print("Missing values distribution:")
print(missing_values_df)
```

```python
df = pd.read_csv('apple_health_merged.csv')

missing_values_count = df.isnull().sum()
missing_values_percentage = 100 * df.isnull().sum() / len(df)

missing_values_df = pd.DataFrame({
    'Missing Count': missing_values_count,
    'Missing Percentage': missing_values_percentage
})

missing_values_df = missing_values_df.sort_values(by='Missing Percentage', ascending=False)

print("Missing values distribution (after reloading data):")
print(missing_values_df)
```

```python
columns_to_drop = missing_values_df[missing_values_df['Missing Percentage'] >= 99].index.tolist()
df_cleaned = df.drop(columns=columns_to_drop)

print(f"Dropped columns: {columns_to_drop}")

# Re-calculate missing values for the cleaned DataFrame
missing_values_count_cleaned = df_cleaned.isnull().sum()
missing_values_percentage_cleaned = 100 * df_cleaned.isnull().sum() / len(df_cleaned)

missing_values_df_cleaned = pd.DataFrame({
    'Missing Count': missing_values_count_cleaned,
    'Missing Percentage': missing_values_percentage_cleaned
})

missing_values_df_cleaned = missing_values_df_cleaned.sort_values(by='Missing Percentage', ascending=False)

print("\nMissing values distribution after dropping columns with >99% missing data:")
print(missing_values_df_cleaned)
```

```python
numerical_cols_to_impute = [
    'steps',
    'distance',
    'blood_glucose',
    'basal_energy',
    'heart_rate',
    'physical_effort'
]

for col in numerical_cols_to_impute:
    if col in df_cleaned.columns:
        median_value = df_cleaned[col].median()
        df_cleaned[col].fillna(median_value, inplace=True)
        print(f"Imputed missing values in '{col}' with median: {median_value:.2f}")

print("\nMissing values after median imputation:")
print(df_cleaned.isnull().sum())
```

```python
numerical_cols_to_impute = [
    'steps',
    'distance',
    'blood_glucose',
    'basal_energy',
    'heart_rate',
    'physical_effort'
]

for col in numerical_cols_to_impute:
    if col in df_cleaned.columns:
        median_value = df_cleaned[col].median()
        df_cleaned[col] = df_cleaned[col].fillna(median_value)
        print(f"Imputed missing values in '{col}' with median: {median_value:.2f}")

print("\nMissing values after median imputation:")
print(df_cleaned.isnull().sum())
```

```python
df_cleaned['timestamp'] = pd.to_datetime(df_cleaned['timestamp'], errors='coerce')
print("Data type of 'timestamp' column after conversion:")
print(df_cleaned['timestamp'].dtype)
```

```python
df_cleaned['year'] = df_cleaned['timestamp'].dt.year
df_cleaned['month'] = df_cleaned['timestamp'].dt.month
df_cleaned['day'] = df_cleaned['timestamp'].dt.day
df_cleaned['day_of_week'] = df_cleaned['timestamp'].dt.dayofweek
df_cleaned['hour'] = df_cleaned['timestamp'].dt.hour

print("First 5 rows of df_cleaned with new date features:")
df_cleaned.head()
```

```python
numerical_cols_for_outliers = [
    'basal_energy',
    'blood_glucose',
    'distance',
    'heart_rate',
    'physical_effort',
    'steps'
]

for col in numerical_cols_for_outliers:
    Q1 = df_cleaned[col].quantile(0.25)
    Q3 = df_cleaned[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df_cleaned[col] = df_cleaned[col].clip(lower=lower_bound, upper=upper_bound)
    print(f"Outliers in '{col}' capped using IQR method.")

print("\nDataFrame head after outlier handling:")
df_cleaned.head()
```
