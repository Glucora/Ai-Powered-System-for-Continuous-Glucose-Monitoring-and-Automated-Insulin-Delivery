```python
# Create a datetime timestamp
cgm["Timestamp"] = pd.to_datetime(
    cgm["Measurement_date"] + " " + cgm["Measurement_time"]
)

# Sort strictly by time
cgm = cgm.sort_values("Timestamp").reset_index(drop=True)

print("First timestamp:", cgm["Timestamp"].iloc[0])
print("Last timestamp:", cgm["Timestamp"].iloc[-1])
```

```python
# Extract glucose values as array
glucose_values = cgm["Measurement"].values.reshape(-1, 1)
```

```python
# Decide train/val/test split (chronological)
train_size = int(len(glucose_values) * 0.7)
val_size   = int(len(glucose_values) * 0.15)
test_size  = len(glucose_values) - train_size - val_size

train_data = glucose_values[:train_size]
val_data   = glucose_values[train_size:train_size+val_size]
test_data  = glucose_values[train_size+val_size:]
```

```python
# Initialize scaler
scaler = MinMaxScaler(feature_range=(0,1))

# Fit ONLY on training data
scaler.fit(train_data)

# Transform all sets
train_scaled = scaler.transform(train_data)
val_scaled   = scaler.transform(val_data)
test_scaled  = scaler.transform(test_data)

# Optional: store scaler for inverse-transform later
```

```python
print("Train min/max:", train_scaled.min(), train_scaled.max())
print("Val min/max:", val_scaled.min(), val_scaled.max())
print("Test min/max:", test_scaled.min(), test_scaled.max())
```

```python
# Inspect duplicate timestamps first
dup_counts = cgm.duplicated(subset=["Timestamp"], keep=False).sum()
print("Number of duplicate rows:", dup_counts)

# Aggregate duplicates by mean
cgm = cgm.groupby("Timestamp", as_index=False)["Measurement"].mean()

# Inspect again
print(cgm.head(10).to_string(index=False))
```

```python
# Calculate differences in minutes
cgm["dt_minutes"] = cgm["Timestamp"].diff().dt.total_seconds() / 60
print("Time difference summary (after duplicates removed):")
print(cgm["dt_minutes"].describe())

# Optional: plot large gaps
print(cgm[cgm["dt_minutes"] > 20][["Timestamp","dt_minutes"]].head())
```

```python
# Choose scaled data (train + val + test concatenated for now)
all_scaled = np.concatenate([train_scaled, val_scaled, test_scaled], axis=0)

L = 24  # lookback
H = 4   # forecast horizon

X, y = [], []

for i in range(len(all_scaled) - L - H + 1):
    x_seq = all_scaled[i:i+L, 0]
    y_seq = all_scaled[i+L:i+L+H, 0]
    # Only append if no NaNs (for gaps)
    if not (np.isnan(x_seq).any() or np.isnan(y_seq).any()):
        X.append(x_seq)
        y.append(y_seq)

X = np.array(X)
y = np.array(y)

print("Windowed data shapes:", X.shape, y.shape)
```

```python
meta = pd.read_csv("patient_1.csv")
gender = meta["Gender"].values[0]  # 0 or 1
gender_array = np.full((X.shape[0], 1), gender)

# Optional: concatenate along feature axis for LSTM
X_lstm = np.expand_dims(X, axis=2)  # shape: (windows, L, 1)
# If adding gender as feature:
X_lstm_with_gender = np.concatenate([X_lstm, np.repeat(gender_array[:, np.newaxis, :], L, axis=1)], axis=2)
print("X shape with gender feature:", X_lstm_with_gender.shape)
```

```python
# Number of sequences
num_samples = X_lstm_with_gender.shape[0]

# Train / val / test sizes
train_size = int(num_samples * 0.7)
val_size   = int(num_samples * 0.15)
test_size  = num_samples - train_size - val_size

print("Number of sequences:")
print(f"Train: {train_size}, Val: {val_size}, Test: {test_size}")
```

```python
# Split X
X_train = X_lstm_with_gender[:train_size]
X_val   = X_lstm_with_gender[train_size:train_size+val_size]
X_test  = X_lstm_with_gender[train_size+val_size:]

# Split y
y_train = y[:train_size]
y_val   = y[train_size:train_size+val_size]
y_test  = y[train_size+val_size:]

# Inspect shapes
print("Shapes after split:")
print("X_train:", X_train.shape, "y_train:", y_train.shape)
print("X_val  :", X_val.shape, "y_val  :", y_val.shape)
print("X_test :", X_test.shape, "y_test :", y_test.shape)
```
